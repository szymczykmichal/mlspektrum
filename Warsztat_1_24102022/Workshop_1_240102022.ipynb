```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

###Stwórzmy przykładowy zestaw danych

```python
X = [-1, 0, 1, 2, 3, 4]
Y = [-3, -1, 1, 3, 5, 7]
```

### Teraz spróbujmy zgadnąć postać funkcji f, która mapuje zbiór X na Y
$y=f(x)$
 

W jaki inny, bardziej pythoniczny sposób możemy wygenerować zbiór wartości funkcji?

```python
Y = []
for x in X:
  y = 2*x-1
  Y.append(y)
```

```python
Y = [2*x-1 for x in X]
```

Zbudujmy model prostej sieci neuronowej, który posiada tylko jeden neuron i spróbujmy nauczyć go postaci funkcji $y$

```python
model = Sequential(
    [
        Dense(units=1, input_shape=[1]),
     ]
)
```

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

Co oznacza zapis "sgd" oraz "mean_squared_error"

Linki:

https://www.tensorflow.org/api_docs/python/tf/keras/losses/MeanSquaredError

https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/SGD

```python
model.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.MeanSquaredError())
```

```python
xs = np.array(X, dtype=float)
ys = np.array(Y, dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

Inny sposób na zdefiniowanie warstwy oraz modelu.

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
model = Sequential([l0])
model.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.MeanSquaredError())
model.fit(xs, ys, epochs=500)
```
