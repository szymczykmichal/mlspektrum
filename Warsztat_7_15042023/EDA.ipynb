```python
### Calculate for states that begin with letter <?> N
```

```python
df.State.unique()
```

```python
def states_starting_w_x(state_col, x="N"):
  if state_col[0] == x:
    return True
  return False
```

```python
df[df["State"].apply(states_starting_w_x)]
```

```python
df[df["State"].apply(lambda state: state[0] == "N")]
```

```python
unique_states = df.State.unique()

list_ = []

for state in unique_states:
  mean_acc_len = df[df["State"] == state]["Account length"].mean()
  list_.append(pd.DataFrame(data = {'State': state, "Mean account length": mean_acc_len}, index=[0]))
  
```

```python
df_acc = pd.concat(list_)
```

```python
df_acc.reset_index(drop=True)
```
